## Train a MLP Model 

You can use the following scripts to train the mlp model, preferably on a GPU machine.

```python
#define the hyperparameters
batch_size    = 32
window_size   =  3 #3 left and 3 right
lr            = 1e-3
device        = "cpu"
epochs        = 4
emb_size      = 200
pretrained    = True
hidden_size   = 128
drop_out      = 0.1
```

```python
train_dataset = MWEDataset("corpus/surf.ftb.train",  isTrain = True,  window_size = window_size)
testset    = MWEDataset("corpus/surf.ftb.test", window_size = window_size)
devset     = MWEDataset("corpus/surf.ftb.dev", window_size = window_size)
print(len(testset.tags_vocab))
```

```python
model = MLPClassifier(
    toks_vocab   = train_dataset.toks_vocab,
    tags_vocab   = train_dataset.tags_vocab, 
    window_size  = window_size,
    emb_size     = emb_size, 
    hidden_size  = hidden_size, 
    drop_out     = drop_out)
```

```python
model.train_model(train_dataset,testset, devset, epochs= epochs, lr=lr, batch_size = batch_size, split_train=0.8)
```

```python
model.save(".trained_models", "mlp_mod.pth")
```

# Train a RNN model

### Vanilla RNN with CRF

```python
#define the hyperparameters
batch_size    = 16
lr            = 1e-3
device        = "cpu"
epochs        = 30
emb_size      = 64
hidden_size   = 64
nb_layers     = 1
drop_out      = 0.1
```

```python
trainset     = rnn_dataset.RnnDataset("corpus/surf.ftb.train",  isTrain = True)
testset      = rnn_dataset.RnnDataset("corpus/surf.ftb.test")
devset       = rnn_dataset.RnnDataset("corpus/surf.ftb.dev")
```

```python
model = rnn_classifier.MweRNN(
    name         = "RNN",
    toks_vocab   = trainset.toks_vocab,
    tags_vocab   = trainset.tags_vocab, 
    emb_size     = emb_size, 
    hidden_size  = hidden_size, 
    drop_out     = drop_out)
```

```python
model.train_model(trainset,testset, devset, epochs= epochs, lr=lr, batch_size = batch_size, split_train=0.8)
```
